```python
# parametric equation of a planar circle.
R = 4.0

def curve(t):

    x = R*np.cos(t)
    y = R*np.sin(t)
    z = 0.0

    return np.array([x, y, z])

def der_curve(t):

    return np.array([-R*np.sin(t), R*np.cos(t), 0.0])

def normalize(vec):

    mag = np.linalg.norm(vec)
    if mag > 0:
        return vec / mag
    else:
        return vec
```

```python
s0 = np.pi/6
p0 = curve(s0)
T0 = normalize(der_curve(s0))
N0  = np.array([-np.cos(s0), -np.sin(s0),   0.0])

origin = gp_Pnt(*p0)
T_dir  = gp_Dir(*T0)   # main axis  → normal to the cross-section plane
N_dir  = gp_Dir(*N0)  

frame  = gp_Ax2(origin, T_dir, N_dir)
```
